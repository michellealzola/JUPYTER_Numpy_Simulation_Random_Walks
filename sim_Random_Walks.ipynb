```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
steps = 10_000
size_of_steps = 0.1
```

```python
random_steps = np.random.normal(loc=0, scale=size_of_steps, size=steps)
```

```python
random_steps
```

```python
paths = np.cumsum(random_steps)
```

```python
paths
```

```python
plt.plot(paths)
plt.title('Simple Random Walks Simulation')
plt.xlabel('Random Steps')
plt.ylabel('Position')
plt.show()
```

```python
# Intermediate level solution
```

```python
num_walks = 100
```

```python
walks = []
for i in range(num_walks):
    random_steps = np.random.normal(loc=0, scale=size_of_steps, size=steps)
    walk = np.cumsum(random_steps)
    walks.append(walk)
```

```python
walks
```

```python
means = np.mean(walks, axis=0)
```

```python
std_dev = np.std(walks, axis=0)
```

```python
for i in range(num_walks):
    plt.plot(walks[i], alpha=0.5)
plt.plot(means, color='blue', linewidth=2)
plt.fill_between(range(steps), means - std_dev, means + std_dev, color='gray', alpha=0.2)
plt.title("Random Walk Intermediate Solution")
plt.xlabel("Random Steps")
plt.ylabel("Positions")
plt.show()
```
